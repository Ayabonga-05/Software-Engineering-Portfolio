# 🧮 Grade Calculator

A Python tool that takes a student's subject scores and generates a full grade report.

**Features:** Average calculation | Letter grades | Best & weakest subject

---

### How to use
Run each cell from top to bottom. The calculator will prompt you for input.


```python
print("="*45)
print("   STUDENT GRADE CALCULATOR")
print("="*45)
```

### Step 1: Get the student name

```python
name = input("Enter Student Name: ")
print(f"Welcome, {name}!")
```

### Step 2: Get the number of subjects

```python
num_subjects = int(input("How many subjects? "))
```

### Step 3: Get subjects names and scores

```python
subjects = []    # List to hold subject names
scores = []      # List to hold subject scores

for i in range(num_subjects):
    subject = input(f"Enter Subject {i + 1} Name: ")
    score = float(input(f"Enter Subject {i + 1} Score: "))
    subjects.append(subject)
    scores.append(score)

print("\nSubjects and Scores recorded!")
```

### Step 4: Calculate the average

```python
average = sum(scores) / len(scores)
average_rounded = round(average, 2)
print(f"Average Score: {average_rounded}%")
```

### Step 5: Get the letter grade

```python
def get_letter_grade(average):
    """Convert average score to letter grade."""
    if average >= 80:
        return "A (Distinction)"
    elif average >= 70:
        return "B (Merit)"
    elif average >= 60:
        return "C (Pass)"
    elif average >= 50:
        return "D (Pass)"
    else:
        return "F (Fail)"

grade = get_letter_grade(average)
print(f"Letter Grade: {grade}")
```

### Step 6: Generate the Final Report

```python
# Find best and weakest subjects
highest = max(scores)
lowest = min(scores)
best_subject = subjects[scores.index(highest)]
weak_subject = subjects[scores.index(lowest)]

# Generate report
print("\n" + "=" * 50)
print(f"  GRADE REPORT FOR: {name.upper()}")
print("=" * 50)

# Display all subjects and scores
for subject, score in zip(subjects, scores):
    print(f"  {subject}: {score}%")

print("-" * 50)
print(f"  Average:           {average_rounded}%")
print(f"  Grade:             {grade}")
print(f"  Best subject:      {best_subject} ({highest}%)")
print(f"  Needs improvement: {weak_subject} ({lowest}%)")
print("=" * 50)
```
