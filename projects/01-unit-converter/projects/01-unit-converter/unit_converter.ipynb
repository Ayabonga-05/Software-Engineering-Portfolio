A Python tool that converts between common units of measurement.

**Covers:** km ↔ miles | °C ↔ °F | kg ↔ lbs

---
### How to use
Run each cell from top to bottom. The last cell launches the converter.

```
# --- Conversion functions ---

def km_to_miles(km):
    return km * 0.621371

def miles_to_km(miles):
    return miles * 1.60934

def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

def fahrenheit_to_celsius(f):
    return (f - 32) * 5/9

def kg_to_pounds(kg):
    return kg * 2.20462

def pounds_to_kg(lbs):
    return lbs / 2.20462

print('Conversion functions loaded.')
```

```
# --- Main program ---

def main():
    print('=' * 40)
    print('     UNIT CONVERTER — by Aya Mthimde')
    print('=' * 40)
    print()
    print('Choose a conversion:')
    print('1. Kilometres  → Miles')
    print('2. Miles       → Kilometres')
    print('3. Celsius     → Fahrenheit')
    print('4. Fahrenheit  → Celsius')
    print('5. Kilograms   → Pounds')
    print('6. Pounds      → Kilograms')
    print('0. Exit')

    while True:
        choice = input('\nEnter choice (0-6): ').strip()

        if choice == '0':
            print('Goodbye!')
            break
        elif choice == '1':
            val = float(input('Enter kilometres: '))
            print(f'  {val} km = {km_to_miles(val):.4f} miles')
        elif choice == '2':
            val = float(input('Enter miles: '))
            print(f'  {val} miles = {miles_to_km(val):.4f} km')
        elif choice == '3':
            val = float(input('Enter Celsius: '))
            print(f'  {val}°C = {celsius_to_fahrenheit(val):.2f}°F')
        elif choice == '4':
            val = float(input('Enter Fahrenheit: '))
            print(f'  {val}°F = {fahrenheit_to_celsius(val):.2f}°C')
        elif choice == '5':
            val = float(input('Enter kilograms: '))
            print(f'  {val} kg = {kg_to_pounds(val):.4f} lbs')
        elif choice == '6':
            val = float(input('Enter pounds: '))
            print(f'  {val} lbs = {pounds_to_kg(val):.4f} kg')
        else:
            print('  Invalid choice. Enter a number between 0 and 6.')

main()
```
