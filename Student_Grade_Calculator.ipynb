# 🧮 Grade Calculator

**By Ayabonga Mthimde** | Software Engineering Portfolio

A Python tool that takes a student's subject scores and generates a full grade report.

**Covers:** Average calculation | Letter grades | Best & weakest subject

---

### How to use
Run each cell from top to bottom. The last cell launches the grade calculator.


```python
print("="*45)
print("   STUDENT GRADE CALCULATOR")
print("="*45)
```

1. Get the student name.

```python
name = input("Enter Student Name: ")
print("Student", name)
```

2. Get the number of subject.

```python
num_subjects = int(input("How many subjects? "))
```

3. Get subjects names and scores.

```python
subjects = []    # empty list, will hold subjects
scores = []    #emty list,will hold subject scores

for i in range(num_subjects):  # repeats once per subject
 subject = input(f"Enter Subject {i + 1} Name:")    #asks for subject name
 score = float(input(f"Enter Subject {i + 1} Score:"))    #asks for subject's score
 subjects.append(subject)    #adds subject to list
 scores.append(score)    #adds score to list

print(subjects)     #print all subjects
print(scores)      #print all scores
```

3. Calculate the avarage.

```python
sum(scores)    #sums up all the scores
len(scores)    #counts items in the list
average = sum(scores)/len(scores)
```

4. Get the letter grade.

```python
def get_letter_grade(average):
    if average >= 80:
        return "A (Distinction)"
    elif average >= 70:
        return "B (Merit)"
    elif average >= 60:
        return "C (Pass)"
    elif average >= 50:
        return "D (Pass)"
    else:
        return "F (Fail)"
grade = get_letter_grade(avarage)
print(grade)
round(average,2)
```

5. The Final Report.

```python
highest = max(scores)                    # finds highest score
lowest = min(scores)                     # finds lowest score
best_subject = subjects[scores.index(highest)]   # finds subject name at same position
weak_subject = subjects[scores.index(lowest)]    # finds subject name at same position

print("=" * 45)
print(f"  REPORT FOR: {name.upper()}")   # .upper() makes name all caps
print("=" * 45)
for subject, score in zip(subjects, scores):     # loops both lists together
    print(f"  {subject}: {score}%")
print("-" * 45)
print(f"  Average:           {round(average, 2)}%")
print(f"  Grade:             {grade}")
print(f"  Best subject:      {best_subject} ({highest}%)")
print(f"  Needs improvement: {weak_subject} ({lowest}%)")
print("=" * 45)
```
